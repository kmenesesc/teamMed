```python
import os
import h5py
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convertDataToPNG(): 
    folders = ["BrainTumorDataSet/brainTumorDataPublic_1-766" ,
           "BrainTumorDataSet/brainTumorDataPublic_767-1532", 
           "BrainTumorDataSet/brainTumorDataPublic_1533-2298", 
           "BrainTumorDataSet/brainTumorDataPublic_2299-3064"]
    current_dir = os.getcwd()

    #Get Directories of Image Folders
    img_dir = []
    for i in range(len(folders)):
        img_dir.append(os.path.join(current_dir, folders[i]))

    #Get Image Paths
    img_files = []
    for i in range(len(folders)):
        img_files.append([os.path.join(img_dir[i],f) for f in os.listdir(img_dir[i]) if (f.endswith('.mat'))]) 
    img_files = np.asarray(img_files).flatten()
    
    #Load Images and Labels
    images = []
    labels = []
    for f in img_files:
        g = h5py.File(f)
        struArray = g[list(g.keys())[0]]
        images.append(struArray['image'][:])
        labels.append(int(struArray['label'][:]))
        
    #Create Directory for Images
    for j in range(max(labels)):
        directory = os.path.join(current_dir, "BrainTumorProcessedImages/"+ str(j+1))
        if not os.path.exists(directory):
            os.makedirs(directory)
    
    #Save Images
    for i in range(len(images)):
        #plt.imshow((images[i]))
        #plt.show()
        cv2.imwrite("BrainTumorProcessedImages/" + str(labels[i]) +"/"+ str(i) + ".png",images[i])
        #Image.fromarray(np.uint8(images[i]), 'L').save("BrainTumorProcessedImages/" + str(labels[i]) +"/"+ str(i) + ".png")
        
    return 0

convertDataToPNG()
```

```python
def splitData(nTrain, nVal, nTest, nLabels): 

    trainPath = []
    validationPath = []
    testPath = []
    
    for j in range(nLabels):
        print(j)
        path = "BrainTumorProcessedImages/" + str(j+1) + "/"
        nFiles = len(os.listdir(path))
        N1 = int(np.floor(nFiles*nTrain))
        N2 = int(np.floor(nFiles*nVal))
        N3 = int(np.floor(nFiles*nTest))
        for i in range(N1): 
            trainPath.append(path + os.listdir(path)[i]+","+str(j+1))
        
        for i in range (N1,N1+N2): 
            validationPath.append(path + os.listdir(path)[i]+","+str(j+1))

        for i in range (N1+N2,N1+N2+N3): 
            testPath.append(path + os.listdir(path)[i]+","+str(j+1))
            
    np.savetxt("Training.txt", trainPath, delimiter=",", fmt="%s")
    np.savetxt("Validation.txt", validationPath, delimiter=",", fmt="%s")
    np.savetxt("Testing.txt", testPath, delimiter=",", fmt="%s")
        
    return 0

splitData(0.8, 0.2, 0, 3)
```
