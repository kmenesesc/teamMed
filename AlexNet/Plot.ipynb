```python
import csv
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read Matlab Results
acc = scipy.io.loadmat("PreTrainedValidation.mat")
acc = acc['acc']
acc = acc[~np.isnan(acc)] #only keep validation accuracies
plt.plot(range(len(acc)),acc,'--o')
plt.title('Validation Accuracy of Pretrained AlexNet')
plt.ylabel('Accuracy')
plt.xlabel('Number of Epochs')
plt.show()

#Read Python Results
file = open('run_tumors_2018-06-15 06_11_23.263597-tag-Accuracy_Validation.csv', 'rb')
reader = csv.reader(file)
has_header = csv.Sniffer().has_header(file.read(1024))
if has_header:
        next(reader)
acc2 = []
for row in reader:
    acc2.append(float(int(float(row[2])*10000))/10000)
plt.plot(range(len(acc2)),acc2,'--o')
plt.title('Validation Accuracy of Original AlexNet')
plt.ylabel('Accuracy')
plt.xlabel('Number of Epochs')
plt.show()
```
